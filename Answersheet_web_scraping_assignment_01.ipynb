# Q 1 
Write a python program to disply all header tags from wikipedia

```python
import pandas as pd
import bs4
import requests
from bs4 import BeautifulSoup 

response=requests.get("https://en.wikipedia.org/wiki/Main_Page")
print("Response_code is",response)
print('_'*30)
soup=BeautifulSoup(response.content)


headers=[]
for i in soup.find_all("span",class_="mw-headline"):
    headers.append(i.text)
    

header_1=[]
for i in soup.find_all("h2",class_="mp-h2"):
    header_1.append(i.text)
# article headers dataframe 1 
art_header=pd.DataFrame({"Header_1":header_1})
    
# all header dataframe 2
header_all=pd.DataFrame({"Header":headers})
```

# Q 2]
Write a python program to dispaly IMDB's top rated 100 movies data

```python
page=requests.get("https://www.imdb.com/list/ls020728848/")
print("Response_code is",page)
print('_'*30)
soup1=BeautifulSoup(page.content)

name=[]
for i in soup1.find_all("h3",class_="lister-item-header"):
    name.append(i.text.split('\n')[2])
    
    
rating=[]
for i in soup1.find_all("div",class_="ipl-rating-star small"):
    rating.append(i.text.split('\n')[-2])
    
year=[]
for i in soup1.find_all("h3",class_="lister-item-header"):
    year.append(i.text.split('\n')[3])
```

```python
top_100_movies=pd.DataFrame({'Movie_names':name,'Rating':rating,'Year_of_release':year},index=range(1,101))
top_100_movies
```

# Q 4] 
Write a python program to display list of respected former Presidents of India

```python
page_pr=requests.get("https://presidentofindia.nic.in/former-presidents.htm")
print("Response_code is",page_pr)
print('_'*30)
soupp=BeautifulSoup(page_pr.content)

pr_name=[]
for i in soupp.find_all("h3"):
    pr_name.append(i.text.split(' (')[0])
    
period=[]
for i in soupp.find_all("p"):
    if ('Term' in i.text):
        period.append(i.text.split("'"))
        
term=[]
for sen in period:
    term.append(str(sen).split(':')[1])

    
Former_President=pd.DataFrame({'Name':pr_name,'Term of Office':term},index=range(1,len(term)+1))
Former_President

    
```

```python
print(len(team))
print(len(matches))
print(len(points))

man_odi_teams=pd.DataFrame({'Team_name':team,'Matches':matches,'Points':points,'Rating':rating},index=range(1,11))
man_odi_teams
```

b) Top 10 ODI Batsman from men,s cricket

```python
site=requests.get("https://www.icc-cricket.com/rankings/mens/player-rankings/odi")
print("response is",site)

code=BeautifulSoup(site.content)

bm=[]
for i in code.find_all("td",class_="table-body__cell name"):
    if len(bm)<9:
        bm.append(i.text.replace("\n",""))
        
country=[]
for i in code.find_all("span",class_='table-body__logo-text'):
    if len(country)<9:
        country.append(i.text)
        
rate=[]
for i in code.find_all("td",class_='table-body__cell u-text-right rating'):
    if len(rate)<9:
        rate.append(i.text)
        
top_bm=code.find("div",class_='rankings-block__banner--name').text

top_country=code.find('div' ,class_="rankings-block__banner--nationality").text.split('\n')[2]

top_rate=code.find('div' ,class_="rankings-block__banner--nationality").text.split('\n')[3].replace(" ",'')
```

```python
# this block of code should executed once only while running first time 

bm.insert(0,top_bm)
country.insert(0,top_country)
rate.insert(0,top_rate)
```

```python
for i in [bm,country,rate]:
    print(len(i))
    
top_odi_batsMen=pd.DataFrame({'player_name':bm,'Team':country,'Rating':rate},index=range(1,11))
top_odi_batsMen    
```

# Q 7]
Write a python program to scrape details from cnbc

```python
req=requests.get("https://www.cnbc.com/world/?region=world")
req
soup=BeautifulSoup(req.content)

head=[]

for i in soup.find_all("a",class_="LatestNews-headline"):
    head.append(i.text)
    
link=[]

for i in soup.find_all("a",class_="LatestNews-headline"):
    link.append(i.get("href")) 
    
tim=[]

for i in soup.find_all("time",class_="LatestNews-timestamp"):
    tim.append(i.text)
    
for i in [head,link,tim]:
    print(len(i))    
```

```python
Latest_News=pd.DataFrame({"Headlines":head,"Time":tim,"News Links":link})
Latest_News
```

# Q 9]
Write python program to scrape details from dineout

```python
reqs=requests.get('https://www.dineout.co.in/delhi-restaurants?search_str=all')
reqs

bea=BeautifulSoup(reqs.content)

#res_name=[]
#for i in bea.find_all("div",class_="restnt-info cursor"):
 #   anch=(i.find_all("a",class_="restnt-name ellipsis"))
  #  for j in anch:
   #     res_name.append(j.text)
        
rest=[]
for i in bea.find_all("a",class_="restnt-name ellipsis"):
    rest.append(i.text)
    
    
cuisine=[]
for i in bea.find_all("span",class_="double-line-ellipsis"):
    cuisine.append(i.text.split('|')[1])
    
loct=[]
for i in bea.find_all("div",'restnt-loc ellipsis'):
    loct.append(i.text)
    
rtn=[]
for i in bea.find_all("div",'restnt-rating rating-4'):
    rtn.append(i.text)
    
img_url=[]
for i in bea.find_all("img",class_="no-img"):
    img_url.append(i.get("data-src"))
    
DineOut=pd.DataFrame({"Restaurant_name":rest,"Cuisine":cuisine,"Location":loct,
              "Rating":rtn,"Image URL":img_url},index=range(1,len(rest)+1))
DineOut    
```

# Q 10]
Write a python program to scrape details of top publications from Google Scholar

```python
req10=requests.get('https://scholar.google.com/citations?view_op=top_venues&hl=en')
req10

soup10=BeautifulSoup(req10.content)

rank=[]
for i in soup10.find_all("td",class_="gsc_mvt_p"):
    rank.append(i.text)
    
pub=[]
for i in soup10.find_all("td",class_="gsc_mvt_t"):
    pub.append(i.text)
    
index=[]
for i in soup10.find_all("a",class_="gs_ibl gsc_mp_anchor"):
    index.append(i.text)
    
median=[]
for i in soup10.find_all("span",class_="gs_ibl gsc_mp_anchor"):
    median.append(i.text)
    
Pub_details=pd.DataFrame({"Publication":pub,"h5-index":index,"h5-median":median},index=rank)  # Rank as  index
Pub_details
```
