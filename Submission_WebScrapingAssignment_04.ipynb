# Q1

```python
dr1=webdriver.Chrome('C:\web driver\chromedriver.exe')
dr1.get('https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos')
dr1.maximize_window()
time.sleep(5)

columns=['Rank','Name','Artist','Views(in Billions)','Upload date']
Dict1={}
for column,index in zip(columns,range(1,6)):
    tag1=dr1.find_elements("xpath","//table[@class='wikitable sortable jquery-tablesorter'][1]/tbody/tr/td[{}]".format(index))
    print(index,' : ',len(tag1))
    Dict1[column]=[]
    for i in tag1:
        Dict1[column].append(i.text)
    
df1=pd.DataFrame(Dict1)
df1


```

```python
df1.to_csv("Most Viewed YouTube video.csv")
dr1.close()
```

# Q2

```python
dr2=webdriver.Chrome('C:\web driver\chromedriverupdated1.exe')
dr2.get('https://www.bcci.tv/')
dr2.maximize_window()
time.sleep(5)

inter=dr2.find_element("xpath","//li[@class='nav-item'][1]")
inter.click()
time.sleep(10)

Dict2={}
Dict2['Match Titles']=[]
Dict2['Series']=[]
Dict2['Places']=[]
Dict2["Date"]=[]
Dict2['Time']=[]

titl_tag=dr2.find_elements("xpath","//div[@class='fix-place ng-binding ng-scope']")
ser_tag=dr2.find_elements("xpath","//h5[@class='fix-text'][2]")
date_tag=dr2.find_elements("xpath","//div[@class='match-card-left match-schedule']")
time_tag=dr2.find_elements("xpath","//div[@class='match-card-right match-schedule ']")
for t,s,d,m in zip(titl_tag,ser_tag,date_tag,time_tag):
    Dict2['Match Titles'].append(t.text.split('-')[0])
    Dict2['Series'].append(s.text)
    Dict2['Places'].append(t.text.split('-')[1])
    Dict2["Date"].append(d.text)
    Dict2["Time"].append(m.text)
    time.sleep(5)
    
df2=pd.DataFrame(Dict2)
df2
```

```python
df2.to_csv("India's International Fixtures.csv")
dr2.close()
```

```python
dr5=webdriver.Chrome('C:\web driver\chromedriverupdated1.exe')
dr5.get('https://github.com/')
dr5.maximize_window()
time.sleep(5)

trend=dr5.find_element("xpath","/html/body/div[1]/header/div/div[2]/nav/ul/li[4]/details/div/ul/li[5]/a")
url5=trend.get_attribute('href')
dr5.get(url5)
time.sleep(10)

title_tag=dr5.find_elements("xpath","//h1[@class='h3 lh-condensed']")
desc_tag=dr5.find_elements("xpath","//p[@class='col-9 color-fg-muted my-1 pr-4']")
lan_tag=dr5.find_elements("xpath","//div[@class='f6 color-fg-muted mt-2']/span[1]")
cont_tag=dr5.find_elements("xpath","//div[@class='f6 color-fg-muted mt-2']/a[2]")
Dict5={}
Dict5['Repository Title']=[]
Dict5['Repository description']=[]
Dict5['Contributors count']=[]
Dict5['Language used']=[]
for t,d,c,l in zip(title_tag,desc_tag,cont_tag,lan_tag):
    Dict5['Repository Title'].append(t.text)
    Dict5['Repository description'].append(d.text)
    Dict5['Contributors count'].append(c.text)
    if 'Built' in l.text:
        Dict5['Language used'].append('-')
    else:
        Dict5['Language used'].append(l.text)
        
df5=pd.DataFrame(Dict5)          
df5
```

```python
dr7=webdriver.Chrome('C:\web driver\chromedriverupdated1.exe')
dr7.get('https://www.naukri.com/naukri-dot-com-recruiters')
dr7.maximize_window()
time.sleep(10)




Dict7={}
Dict7["Name"]=[]
Dict7["Designation"]=[]
Dict7["Company"]=[]
Dict7["Skills they hire for"]=[]
Dict7["Location"]=[]
name7=dr7.find_elements("xpath","//p[@class='highlightable']/a/span")
desg7=dr7.find_elements("xpath","//span[@class='ellipsis clr']")
comp7=dr7.find_elements("xpath","//p[@class='highlightable']/a[2]")
skill7=dr7.find_elements("xpath","//div[@class='hireSec highlightable']")
loc7=dr7.find_elements("xpath","//p[@class='highlightable']")

for  n,d,c,s,l in zip(name7,desg7,comp7,skill7,loc7):
    Dict7["Name"].append(n.text)
    Dict7["Designation"].append(d.text)
    Dict7["Company"].append(c.text)
    Dict7["Skills they hire for"].append(s.text)
    try:
        Dict7["Location"].append(l.text.split('\n')[3])
    except:
        Dict7["Location"].append('-')

df7=pd.DataFrame(Dict7)    
df7
```

```python
dr9=webdriver.Chrome('C:\web driver\chromedriverupdated1.exe')
dr9.get('https://www.imdb.com/list/ls095964455/')
dr9.maximize_window()
time.sleep(10)

Dict9={}
Dict9["Name"]=[]
Dict9["Year span"]=[]
Dict9["Genre"]=[]
Dict9["Run time"]=[]
Dict9["Ratings"]=[]
Dict9["Votes"]=[]


name=dr9.find_elements("xpath","//h3[@class='lister-item-header']/a")
span=dr9.find_elements("xpath","//h3[@class='lister-item-header']/span[2]")
genre=dr9.find_elements("xpath","//div[@class='lister-item-content']/p[1]/span[5]")
run=dr9.find_elements("xpath","//div[@class='lister-item-content']/p[1]/span[3]")
rating=dr9.find_elements("xpath","//div[@class='ipl-rating-star small']/span[2]")
votes=dr9.find_elements("xpath","//div[@class='lister-item-content']/p[4]/span[2]")

for n,s,g,u,r,v in zip(name,span,genre,run,rating,votes):
    Dict9["Name"].append(n.text)
    Dict9["Year span"].append(s.text)
    Dict9["Genre"].append(g.text)
    Dict9["Run time"].append(u.text)
    Dict9["Ratings"].append(r.text)
    Dict9["Votes"].append(v.text)
    
df9=pd.DataFrame(Dict9)
df9
```
