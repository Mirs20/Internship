```python
# Q11] Write a python program to find the factorial of a number.

num = int(input("Enter a number: "))    
a = 1    
if num < 0:    
    print(" Negative numbers have not factorial ")    
elif num == 0:    
    print("The factorial of 0 is 1")    
else:    
    for i in range(1,num + 1):    
        a = a*i    
    print("The factorial of",num,"is",a)    
```

```python
#Q 12] Write a python program to find whether a number is prime or composite.

def Prime_Check(num):  
    if num > 1:  
        for i in range(2, int(num/2)+1):  
            if (num % i) == 0:  
                print(num, "is composite  number")  
                break    
        else:  
                print(num, "is a prime number")    
    else:  
        print(num, "is neither composite nor prime number")  
num = int(input("Enter an input number:"))  
  
Prime_Check(num)  
```

```python
# Q 13] Write a python program to check whether a given string is palindrome or not.


string=input('Enter a string or number')


def Pal_check(string):
    return  ''.join(reversed(string))

if (string == Pal_check(string)):
    print( string,'is Palindrome ')
else:
    print(string,'is not palindrome')
```
