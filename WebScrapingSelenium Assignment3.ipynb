```python
dict1={}
dict1['Brand']=[]
dict1['Model']=[]
dict1['Price']=[]
dict1["Expected Delivery"]=[]
dict1['Return/Exchange']=[]

for i in urls:
    drp=webdriver.Chrome('C:\web driver\chromedriver.exe')
    drp.get(str(i))
    drp.maximize_window()
    time.sleep(5)
   
    try:
        mod_tag=drp.find_element_by_xpath("//div[@id='productOverview_feature_div']/div/table/tbody/tr[1]/td[2]")
        dict1['Model'].append(mod_tag.text)
    except:
        dict1['Model'].append('-')

    try:
        brand_tag=drp.find_element_by_xpath("//div[@id='productOverview_feature_div']/div/table/tbody/tr[3]/td[2]")
        dict1['Brand'].append(brand_tag.text)
    except:
        dict1['Brand'].append('-')

     
    price_tag=drp.find_element_by_xpath("//div[@id='apex_desktop']/div/div")
    
    if price_tag.text is None:
        dict1['Price'].append('-')
    elif 'M.R.P.' in price_tag.text:
        dict1['Price'].append(price_tag.text.split('\n')[3])
    elif "%" in price_tag.text:
        dict1['Price'].append(price_tag.text.split('\n')[1])
    else:
        dict1['Price'].append(price_tag.text.split('\n')[0])
    
    try:
    
        del_tag=drp.find_element_by_xpath("//div[@id='mir-layout-DELIVERY_BLOCK-slot-PRIMARY_DELIVERY_MESSAGE_LARGE']/span/span")
        dict1["Expected Delivery"].append(del_tag.text)
    except:
        dict1['Expected Delivery'].append("-")

   
    try:
        
        WebDriverWait(drp,1).until(EC.presence_of_element_located((By.XPATH,"//div[@id='RETURNS_POLICY']")))
        ret_tag=drp.find_element_by_xpath("//div[@id='RETURNS_POLICY']") 
        dict1['Return/Exchange'].append(ret_tag.text)
        
    except:
        dict1["Return/Exchange"].append('_')
    
    drp.close()
```

```python
dr8.get("http://www.forbes.com/")
dr8.maximize_window()
```

```python
top8=dr8.find_element_by_xpath("/html/body/div[1]/header/nav/div[1]/button[1]")
top8.click()
time.sleep(10)
```

```python
bil8=dr8.find_element_by_xpath("//li[@class='header__section']/a")
dr8.get(bil8.get_attribute("href"))
time.sleep(10)
```

```python
WebDriverWait(dr8, 10).until(EC.visibility_of_element_located((By.XPATH,("//div[@class='rank']"))))
rank_tag=dr8.find_elements_by_xpath("//div[@class='rank']")
len(rank_tag)
```

```python
ranks=[]
names=[]
worths=[]
country=[]
age=[]
source=[]
indust=[]

rank_tag=dr8.find_elements_by_xpath("//div[@class='rank']")
name_tag=dr8.find_elements_by_xpath("//div[@class='personName']")
worth_tag=dr8.find_elements_by_xpath("//div[@class='netWorth']")
con_tag=dr8.find_elements_by_xpath("//div[@class='countryOfCitizenship']")
age_tag=dr8.find_elements_by_xpath("//div[@class='age']")
ind_tag=dr8.find_elements_by_xpath("//div[@class='category']")
source_tag=dr8.find_elements_by_xpath("//span[@class='source-text']")

for i,n,w,c,a,s,j in zip(rank_tag,name_tag,worth_tag,con_tag,age_tag,source_tag,ind_tag):
    
    ranks.append(i.text)
    names.append(n.text)
    worths.append(w.text)             
    country.append(c.text)             
    age.append(a.text)
    source.append(s.text)             
    indust.append(j.text)             
                 
                 
  
```

```python
len(ranks)==len(names)==len(age)==len(worths)==len(country)==len(indust)==len(source)
```

```python
df8=pd.DataFrame({"Rank":ranks,"Name":names,"Net Worth":worths,"Age":age,"Citizenship":country,"Source":source,"Industry":indust})
df8
```

```python
time.sleep(5)
dict10={}
dict10['Hostel Name']=[]
dict10['Distance from city centre']=[]
dict10['Property Description']=[]
dict10['Rating']=[]
dict10['Overall Review']=[]
dict10['Dorms from Price']=[]
dict10['Total Reviews']=[]
dict10['Privates from Price']=[]
dict10['facilities']=[]



    

name_tag=dr10.find_elements_by_xpath("//h2[@class='title title-6']")
for i in name_tag:
    dict10['Hostel Name'].append(i.text)

dist_tag=dr10.find_elements_by_xpath("//span[@class='description']")
for i in dist_tag:
    dict10['Distance from city centre'].append(i.text.split('-')[1].split(' ')[1])



desc_tag=dr10.find_elements_by_xpath("//div[@class='rating-factors prop-card-tablet rating-factors small']")
for i in desc_tag:
    if i.text=='':
        dict10['Property Description'].append('-')
    else:
        dict10['Property Description'].append(i.text.replace('\n',','))


r_tag=dr10.find_elements_by_xpath("//div[@class='rating rating-summary-container big']")

for i in r_tag:
    if 'No Rating' in i.text:
        dict10['Rating'].append("-")
    else:
        dict10['Rating'].append(i.text.split('\n')[0])

for i in r_tag:
    if "Rating" in i.text:
        dict10['Overall Review'].append("-")
    else:
        dict10['Overall Review'].append(i.text.split('\n')[1])

for i in r_tag:
    dict10['Total Reviews'].append(i.text.split('\n')[-1].split(' ')[0])







private_tag=dr10.find_elements_by_xpath("//div[@class='price-col'][1]")
for i in private_tag:
    if 'Rs' in i.text:
        dict10['Privates from Price'].append(i.text.split('\n')[1].split(' ')[-1])
    else:
        dict10['Privates from Price'].append(i.text)


dorms_tag=dr10.find_elements_by_xpath("//div[@class='price-col'][2]")

for i in dorms_tag:
    if 'Rs' in i.text:
        dict10['Dorms from Price'].append(i.text.split('\n')[1].split(' ')[-1])
    else:
        dict10['Dorms from Price'].append(i.text)



fac_tag=dr10.find_elements_by_xpath("//div[@class='facilities-label facilities']")
for i in fac_tag:
    if i.text=='':
        dict10['facilities'].append('-')
    else:
        dict10['facilities'].append(i.text.replace('\n',','))

```
